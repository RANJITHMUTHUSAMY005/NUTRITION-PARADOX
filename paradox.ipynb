# **COUNTRY CODES INTO FULLNAME**

```python
!pip install pycountry
```

```python
import pycountry
special_cases = {
    'GLOBAL': 'Global',
    'WB_LMI': 'Low & Middle Income',
    'WB_HI': 'High Income',
    'WB_LI': 'Low Income',
    'EMR': 'Eastern Mediterranean Region',
    'EUR': 'Europe',
    'AFR': 'Africa',
    'SEAR': 'South-East Asia Region',
    'WPR': 'Western Pacific Region',
    'AMR': 'Americas Region',
    'WB_UMI': 'Upper Middle Income'
}
def get_country_full_name(code):
    if code in special_cases:
        return special_cases[code]
    country = pycountry.countries.get(alpha_3=code)
    if country:
        return country.name
    else:
        return None
df_mal['Country_name'] = df_mal['Country'].apply(get_country_full_name)
df_obesity['Country_name'] = df_obesity['Country'].apply(get_country_full_name)
```

```python
df_obesity['CI_Width'] = df_obesity['UpperBound'] - df_obesity['LowerBound']
```

```python
df_mal['CI_Width'] = df_mal['UpperBound'] - df_mal['LowerBound']
```

```python
def categorize_obesity(Mean_Estimate):
  if Mean_Estimate >= 30:
    return 'High'
  elif 25 >= Mean_Estimate <= 29.9:
    return 'Moderate'
  else:
    return 'Low'
df_obesity['obesity_level']=df_obesity['Mean_Estimate'].apply(categorize_obesity)
```

```python
df_mal['Mean_Estimate'].value_counts()
```

```python
def categorize_malnut(Mean_Estimate):
  if Mean_Estimate >= 20:
    return 'High'
  elif 10 >= Mean_Estimate <= 19.9:
    return 'Moderate'
  else:
    return 'Low'
df_mal['Malnutrition_level']=df_mal['Mean_Estimate'].apply(categorize_malnut)
```

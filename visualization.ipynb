```python
import pandas as pd
df_obesity = pd.read_csv('obesity.csv')
df_obesity
```

```python
df_mal = pd.read_csv('malnutrition.csv')
df_mal
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px #importing plotly for graphs
import plotly.graph_objects  as go

# Select only numeric columns from the DataFrame
numeric_df = df_mal.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix for the numeric columns
corr_table = numeric_df.corr()

# Display the correlation matrix
print(corr_table)

plt.figure(figsize=(10, 8))
sns.heatmap(corr_table, annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
numeric_df_1 = df_obesity.select_dtypes(include=['float64', 'int64'])
for i in numeric_df_1.columns:
  sns.kdeplot(numeric_df_1[i])
  plt.show()
```

```python
numeric_df_2 = df_mal.select_dtypes(include=['float64', 'int64'])
for i in numeric_df_2.columns:
  sns.kdeplot(numeric_df_2[i])
  plt.show()
```

```python
df_mal_grouped = df_mal.groupby('Region', as_index=False)['Mean_Estimate'].mean()
plt.figure(figsize=(10, 6))
sns.barplot(data=df_mal_grouped, x='Region', y='Mean_Estimate', ci=None, palette='viridis')
plt.title('Average Malnutrition Levels by Region')
plt.xlabel('Region')
plt.ylabel('Mean Malnutrition Estimate')
plt.xticks(rotation=90)
plt.show()
```

```python
df_obese_grouped = df_obesity.groupby('Region', as_index=False)['Mean_Estimate'].mean()
plt.figure(figsize=(10, 6))
sns.barplot(data=df_mal_grouped, x='Region', y='Mean_Estimate', ci=None, palette='viridis')
plt.title('Average obesity Levels by Region')
plt.xlabel('Region')
plt.ylabel('Mean obesity Estimate')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(5, 5))
obesity_level_counts = df_obesity['obesity_level'].value_counts()
plt.pie(obesity_level_counts, labels=obesity_level_counts.index, autopct='%1.1f%%', colors=['lightgreen', 'lightblue', 'salmon'])
plt.title('Distribution of Obesity Levels')
plt.show()
```

```python
plt.figure(figsize=(5,5))
malnutrition_level_counts = df_mal['Malnutrition_level'].value_counts()
plt.pie(malnutrition_level_counts, labels=malnutrition_level_counts.index, autopct='%1.1f%%',colors= None)
plt.title('Distribution of Malnutrition Levels')
plt.show()
```

```python
plt.hist(numeric_df_1['Mean_Estimate'], bins=30)
plt.xlabel("Mean_Estimate")
plt.ylabel("Frequency")
plt.title("Distribution of Obesity Mean Estimate")
plt.show()
```

```python
plt.hist(numeric_df_2['Mean_Estimate'], bins=30)
plt.xlabel("Mean_Estimate")
plt.ylabel("Frequency")
plt.title("Distribution of malnutrition Mean Estimate")
plt.show()
```

# **Trends in Obesity Estimates Over Time**

```python
plt.figure(figsize=(20, 8))
sns.lineplot(data=df_obesity, x='Year', y='Mean_Estimate', hue='Region', marker='o', palette='tab20')
plt.title('Trends in Obesity Estimates Over Time by Region')
plt.xlabel('Year')
plt.ylabel('Mean Obesity Estimate')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(20, 8))
sns.lineplot(data=df_mal, x='Year', y='Mean_Estimate', hue='Region', marker='o', palette='tab20')
plt.title('Trends in Obesity Estimates Over Time by Region')
plt.xlabel('Year')
plt.ylabel('Mean Obesity Estimate')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
